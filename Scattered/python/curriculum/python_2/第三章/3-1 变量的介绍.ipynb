# 变量的命名规则

# 1.变量只能包含字母、数字和下划线。
# 变量名可以字母或下划线开头，但绝不能是数字开头。比如说，可以将变量命名为 data_1，但绝不能命名为 1_data。

```python
data_1 = 2
```

```python
1_data = 2
```

```python
_data = 2
```

# 2.不能将python关键字作为变量名。

```python
from IPython.display import Image
Image(filename = 'python 关键字.png')
```

```python
None = 3
```

# 3.变量名最好精简又具有描述性。如student_name 比 s_n好，fruits 比 f好。
# 4.变量名最好使用小写字母，不建议使用大写字母命名
